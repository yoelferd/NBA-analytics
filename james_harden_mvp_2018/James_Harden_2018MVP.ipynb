```python
def changePosition(x):
    if ("SG" in x) or ("PG" in x):
        return "Guard"
    elif ("SF" in x) or ("PF" in x):
        return "Forward"
    else:
        return "Center"
```

```python
eighteen['Position'] = eighteen['Pos'].apply(changePosition)
```

```python
mask1 = (eighteen['Player'] == 'Lou Williams') & (eighteen['Tm'] != 'TOT')
mask2 = (eighteen['Player'] == 'DeMarcus Cousins') & (eighteen['Tm'] != 'TOT')
eighteen = eighteen[~(mask1 | mask2)]
```

```python
def hollingerMVP(x):
    if x >= 35.0:
        return '1. All-Time Great Season'
    elif x >= 30.0 and x < 35:
        return '2. Runaway MVP Candidate'
    elif x >= 27.5 and x < 30:
        return '3. Strong MVP Candidate'
    elif x >= 25.0 and x < 27.5:
        return '4. Weak MVP Candidate'
    else:
        return "Not MVP Candidate"
```

```python
eighteen['MVP Candidacy'] = eighteen['PER'].apply(hollingerMVP)
```

```python
mvpCandidates = eighteen[eighteen['MVP Candidacy'] != 'Not MVP Candidate']
```

```python
mvpCandidates = mvpCandidates.sort_values("MVP Candidacy").reset_index()
mvpCandidates.drop(["level_0","index"],axis=1,inplace=True)
```

```python
for count,player in enumerate(mvpCandidates['Player']):
    print (player,"had a PER of",mvpCandidates["PER"][count],"and is considered a",mvpCandidates['MVP Candidacy'][count],"by Hollingers function")
```
