```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from plotly.offline import init_notebook_mode,iplot,iplot_mpl,download_plotlyjs,enable_mpl_offline
import plotly.graph_objs as go
import os

%matplotlib inline

init_notebook_mode(connected=True)
```

```python
players = pd.read_csv("nba-players-stats/Players.csv",usecols=["Player","height","weight","collage","born","birth_city","birth_state"])
players[["height","weight","born"]] = players[["height","weight","born"]].fillna(0).astype(int)
season = pd.read_csv("nba-players-stats/Seasons_Stats.csv")
season[["Year","Age","G","GS","AST","STL","BLK","PTS","PF","TOV","FT","FTA","2P","2PA","3P","3PA","ORB","TRB","DRB"]] = season[["Year","Age","G","GS","AST","STL","BLK","PTS","PF","TOV","FT","FTA","2P","2PA","3P","3PA","ORB","TRB","DRB"]].fillna(0).astype(int)
season = season[season["Year"] >= 2009]
season["PPG"] = season["PTS"] / season["G"]
season["APG"] = season["AST"] / season["G"]
season["3PG"] = season["3P"] / season["G"]
season["RPG"] = season["TRB"] / season["G"]
season["FTPG"] = season["FTA"] / season["G"]
season["TOPG"] = season["TOV"] / season["G"]
season["PPG"] = round(season["PPG"],1)
```

```python
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
stats = season[["Year","Player","Pos","Age","Tm","PER","PTS","PPG","APG","3PG","RPG","FTPG","TOPG","FG%","3P%","2P%","FT%",'WS','BPM','OWS','DWS']].reset_index()
```

```python
eighteen = stats[(stats['Year'] == 2018) & (stats['PPG'] >= 15)]
```

```python
#adding in this season

advanced = pd.read_csv("advplayerdata.csv")
totals = pd.read_csv("playertotals.csv")

advanced[['GS', 'FG', 'FGA', 'FG%', '3P', '3PA', '3P%', '2P', '2PA', '2P%', 'eFG%', 'FT', 'FTA', 'FT%', 'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS']]=totals[['GS', 'FG', 'FGA', 'FG%', '3P', '3PA', '3P%', '2P', '2PA', '2P%', 'eFG%', 'FT', 'FTA', 'FT%', 'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS']].fillna(0).astype(int)

#['PER', 'TS%', '3PAr', 'FTr', 'ORB%', 'DRB%', 'TRB%', 'AST%', 'STL%', 'BLK%', 'TOV%', 'USG%', 'Unnamed: 19', 'OWS', 'DWS', 'WS', 'WS/48', 'Unnamed: 24', 'OBPM', 'DBPM', 'BPM', 'VORP']
#['GS', 'FG', 'FGA', 'FG%', '3P', '3PA', '3P%', '2P', '2PA', '2P%', 'eFG%', 'FT', 'FTA', 'FT%', 'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS']
season18 = advanced

season18[["Age","G","GS","AST","STL","BLK","PTS","PF","TOV","FT","FTA","2P","2PA","3P","3PA","ORB","TRB","DRB"]] = season18[["Age","G","GS","AST","STL","BLK","PTS","PF","TOV","FT","FTA","2P","2PA","3P","3PA","ORB","TRB","DRB"]].fillna(0).astype(int)

season18["PPG"] = season18["PTS"] / season18["G"]
season18["APG"] = season18["AST"] / season18["G"]
season18["3PG"] = season18["3P"] / season18["G"]
season18["RPG"] = season18["TRB"] / season18["G"]
season18["FTPG"] = season18["FTA"] / season18["G"]
season18["TOPG"] = season18["TOV"] / season18["G"]
season18["PPG"] = round(season18["PPG"],1)
season18["Year"] = [2018]*len(season18)
litty = pd.concat([season, season18], ignore_index=True, sort=False)
litty
```

```python
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
stats = litty[["Year","Player","Pos","Age","Tm","PER","PTS","PPG","APG","3PG","RPG","FTPG","TOPG","FG%","3P%","2P%","FT%",'WS','BPM','OWS','DWS']].reset_index()
```

```python
eighteen = stats[(stats['Year'] == 2018) ]#& (stats['PPG'] >= 15)]
mask1 = (eighteen['Player'] == 'Lou Williams') & (eighteen['Tm'] != 'TOT')
mask2 = (eighteen['Player'] == 'DeMarcus Cousins') & (eighteen['Tm'] != 'TOT')
eighteen = eighteen[~(mask1 | mask2)]
```

```python
def changePosition(x):
    if ("SG" in x) or ("PG" in x):
        return "Guard"
    elif ("SF" in x) or ("PF" in x):
        return "Forward"
    else:
        return "Center"
```

```python
eighteen['Position'] = eighteen['Pos'].apply(changePosition)
eighteen
```

```python
# fig = plt.figure(figsize=(12,8))
# ax = fig.add_axes([0,0,.7,.7])
# ax.set_title('WS vs. Pts')
trace = go.Scatter(x = eighteen['WS'],y = eighteen['PPG'],mode = 'markers+text',text=eighteen['Player'],textposition = 'middle right')

data = [trace]
layout = dict(title = 'WS vs. PPG',
              hovermode = 'closest',
              yaxis = dict(zeroline = False,title="PPG"),
              xaxis = dict(zeroline = False,range=(0,25),title="Win Share")
             )

fix = dict(data=data,layout=layout)

iplot(fix,filename='basic-scatter')
```
